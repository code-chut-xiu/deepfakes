```python
from PIL import Image
import os

def convert_tif_to_jpg(source_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)

    for file in os.listdir(source_dir):
        if file.lower().endswith(".tif") or file.lower().endswith(".tiff"):
            tif_path = os.path.join(source_dir, file)
            jpg_name = os.path.splitext(file)[0] + ".jpg"
            jpg_path = os.path.join(output_dir, jpg_name)

            try:
                img = Image.open(tif_path).convert("RGB")  # Convert to RGB
                img.save(jpg_path, "JPEG", quality=95)
                print(f"Converted: {file} → {jpg_name}")
            except Exception as e:
                print(f"Failed to convert {file}: {e}")

# Example usage
convert_tif_to_jpg("/Users/Hung.Le/Downloads/CASIA2/tiff", "/Users/Hung.Le/Downloads/CASIA2/tif-to-jpg")
```
